```
import random

print("############################# WELCOME TO SNAKES AND LADDERS GAME #############################")

Player1 = input("Please enter a name for first player: ")
Player2 = input("Please enter a name for second player: ")
print("\nMatch will be played between '" + Player1 + "' and '" + Player2 + "'\n")


position1=0
position2=0
print("The initial position of '" +Player1+ "'is'",position1)
print("The initial position of '" +Player2+ "'is'",position2)


#assigning the positions of the snades and ladders
snakes={
17 : 7,
54 : 34, 
62 : 19, 
98 : 79
}
ladders={
3 : 38, 
24 : 33, 
42 : 93, 
72 : 84
}
snakes_keys=list(snakes.keys())
ladders_keys=list(ladders.keys())
def turn(x,pos):
    pos=pos+x
    if pos in snakes:
        pos=snakes[pos]
        print("Oh no!..You got a snake bite")
    elif pos in ladders:
        pos=ladders[pos]
        print("Hurray!!, you climb a stair")
    return pos


player_turn=1


while position1 < 100 and position2 < 100:
    if(player_turn==1):
        Player1_chance=input("Player 1: ")  
        if(Player1_chance=="quit"):     #if player1 quits the game then automatically player 2 wins the game 
            print("Player 1 quits.")
            print("Player 2 won the game")
            break
        elif(Player1_chance=="roll"):       #if player1 wants to roll the dice then random number will be generated in the dice and it will get added to the current position
            Player1_chance=random.randint(1, 6)
            print("You got a ",Player1_chance)
            if position1+Player1_chance>100:
                print("Invalid Move :( Try Again")   #if the current position of the player and the value he gpt in the dice is > 100 he can't move from the current position.
                continue
            else:

                position1=turn(Player1_chance,position1)
                print(" Your final position is:",position1)
                player_turn=2
                if(position1==100):
                    print("Player 1 won the game")

        else:
            Player1_chance=int(Player1_chance)
            if( Player1_chance<1 or Player1_chance >20):  #the player must input a number in between 1-20 when he/she got a chance to play
                print("Please Enter Valid Choice- Between 1-20. Try Again")
                continue
            else:
                print("You got a ",Player1_chance)
                if position1+Player1_chance>100:
                    print("Invalid Move :( Try Again")
                    continue
                else:
                    position1=turn(Player1_chance,position1)
                    print(" Your final position is:",position1)
                    player_turn=2
                    
                    if(position1==100):
                        print("Player 1 won the game")



    else:
        Player2_ch=input("Player 2: ")
        if(Player2_ch=="quit"):
            print("Player 2 quits.")
            print("Player 1 won the game")
            break
        elif(Player2_ch=="roll"):
            Player2_chance=random.randint(1, 6)
            print("You got a ",Player2_chance)
            if position2+Player2_chance>100:
                print("Invalid Move :( Try Again")
                continue
            else:

                position2=turn(Player2_chance,position2)
                print(" Your final position is:",position2)
                player_turn=1
                
                if(position2==100):
                    print("Player 2 won the game")

        else:
            Player2_chance=int(Player2_ch)
            if( Player2_chance<1 or Player2_chance >20):
                print("Please Enter Valid Choice- Between 1-20. Try Again")
                continue
            else:
                print("You got a ",Player2_chance)
                if position2+Player2_chance>100:
                    print("Invalid Move :( Try Again")
                    continue
                else:
                    position2=turn(Player2_chance,position2)
                    print(" Your final position is:",position2)
                    player_turn=1
                    
                    if(position2==100):
                        print("Player 2 won the game")

print("############################# Game Successfully Finished #############################")
```
